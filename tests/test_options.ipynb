## 1. Non-negative Random Variable
**Goal:** Show that $\mathbb{E}[X] = \int_{0}^{T} \mathbb{P}(X > x) \, dx$.

**Proof:**
Using the indicator function $I_{\{X > x\}}$ and **Fubini's Theorem** to swap the expectation and integral:

$$
\begin{aligned}
\int_{0}^{T} \mathbb{P}(X > x) \, dx &= \int_{0}^{T} \mathbb{E}[I_{\{X > x\}}] \, dx \\
&= \mathbb{E} \left[ \int_{0}^{T} I_{\{X > x\}} \, dx \right] \\
&= \mathbb{E} \left[ \int_{0}^{X} 1 \, dx \right] \quad (\text{assuming } T \ge X) \\
&= \mathbb{E}[X]
\end{aligned}
$$

---

## 2. General Random Variable
**Goal:** Show that $\mathbb{E}[X] = \int_{0}^{T} \mathbb{P}(X > x) \, dx - \int_{-\infty}^{0} \mathbb{P}(X < x) \, dx$.

**Proof:**
Decompose $X$ into its positive and negative parts: $X = X^+ - X^-$, where $X^+ = \max(0, X)$ and $X^- = \max(0, -X)$. By linearity, $\mathbb{E}[X] = \mathbb{E}[X^+] - \mathbb{E}[X^-]$.

1. **For $X^+$:** Applying the result from (1):
   $$\mathbb{E}[X^+] = \int_{0}^{T} \mathbb{P}(X^+ > x) \, dx = \int_{0}^{T} \mathbb{P}(X > x) \, dx$$

2. **For $X^-$:** Since $X^-$ is non-negative:
   $$\mathbb{E}[X^-] = \int_{0}^{\infty} \mathbb{P}(X^- > x) \, dx = \int_{0}^{\infty} \mathbb{P}(X < -x) \, dx$$
   Perform a change of variables $t = -x$ (hence $dt = -dx$):
   $$\mathbb{E}[X^-] = \int_{0}^{-\infty} \mathbb{P}(X < t) (-dt) = \int_{-\infty}^{0} \mathbb{P}(X < t) \, dt$$

3. **Conclusion:**
   Substituting parts 1 and 2 back into $\mathbb{E}[X] = \mathbb{E}[X^+] - \mathbb{E}[X^-]$:
   $$\mathbb{E}[X] = \int_{0}^{T} \mathbb{P}(X > x) \, dx - \int_{-\infty}^{0} \mathbb{P}(X < x) \, dx$$

## 2.1 Mollification

### 2.1.1 Show Convergence
**Goal:** Show that $\varphi_\epsilon(x) \to (x - K)^+$ as $\epsilon \downarrow 0$.

**Proof:**
* **Case $x \le K$:** As $\epsilon \downarrow 0$, the exponent $(x-K)/\epsilon \to -\infty$ (or $0$ if $x=K$), so $e^{(x-K)/\epsilon} \to 0$ (or $1$). 
    $$\varphi_\epsilon(x) = \epsilon \log(1 + e^{(x-K)/\epsilon}) \to 0 \cdot \log(1) = 0$$
* **Case $x > K$:** Rewrite the function by factoring out the exponential term:
    $$\varphi_\epsilon(x) = \epsilon \log \left( e^{(x-K)/\epsilon} (1 + e^{-(x-K)/\epsilon}) \right) = (x-K) + \epsilon \log(1 + e^{-(x-K)/\epsilon})$$
    As $\epsilon \downarrow 0$, the exponent $-(x-K)/\epsilon \to -\infty$, causing $e^{-(x-K)/\epsilon} \to 0$. The second term vanishes:
    $$\lim_{\epsilon \downarrow 0} \varphi_\epsilon(x) = (x-K) + 0 \cdot \log(1) = x-K$$

**Conclusion:** $\lim_{\epsilon \downarrow 0} \varphi_\epsilon(x) = \max(0, x-K) = (x-K)^+$.

---

### 2.1.2 Compute the First Four Derivatives
Let $y = \frac{x-K}{\epsilon}$. The first derivative simplifies to the standard logistic sigmoid function, let's call it $S(y) = \frac{1}{1+e^{-y}}$. 
Note that $\frac{dy}{dx} = \frac{1}{\epsilon}$ and the derivative of the sigmoid is $S'(y) = S(y)(1-S(y))$.

Using the chain rule repeatedly:

1.  **First Derivative:**
    $$\varphi_\epsilon'(x) = S(y) = \frac{1}{1 + e^{-(x-K)/\epsilon}}$$
2.  **Second Derivative:**
    $$\varphi_\epsilon''(x) = \frac{1}{\epsilon} S(y)(1-S(y))$$
3.  **Third Derivative:**
    $$\varphi_\epsilon'''(x) = \frac{1}{\epsilon^2} S(y)(1-S(y))(1-2S(y))$$
4.  **Fourth Derivative:**
    $$\varphi_\epsilon^{(4)}(x) = \frac{1}{\epsilon^3} S(y)(1-S(y))(1-6S(y)+6S(y)^2)$$

---

### 2.1.3 Interpret the Limits as $\epsilon \downarrow 0$
As the smoothing parameter $\epsilon \to 0$, these derivatives correspond to the limit distributions of the option's "Greeks":

* **1st Derivative (Delta):** It represents the option's delta, shifting abruptly from $0$ (out-of-the-money) to $1$ (in-the-money) at maturity.
* **2nd Derivative (Gamma):** This shows that at maturity, all Gamma  is infinitely concentrated exactly at the strike price $K$.
* **3rd Derivative (Speed):** It indicates the extreme, instantaneous reversal of Gamma right at the strike price.
* **4th Derivative:** It representing even higher-order rate changes concentrated at the strike.

## 2.1 Expansion

Let:
$$\mu = \mathbb{E}[X].$$

Perform a **Taylor expansion** of $\varphi(X)$ at $\mu$:
$$\varphi(X) = \sum_{n=0}^{\infty} \frac{\varphi^{(n)}(\mu)}{n!}(X - \mu)^n.$$

Take the **expectation** of both sides:
$$V = \mathbb{E}[\varphi(X)] = \sum_{n=0}^{\infty} \frac{\varphi^{(n)}(\mu)}{n!} \mathbb{E}[(X - \mu)^n].$$

Let the **$n$-th order central moment** be:
$$m_n = \mathbb{E}[(X - \mu)^n],$$

then:
$$V = \sum_{n=0}^{\infty} \frac{\varphi^{(n)}(\mu)}{n!} m_n.$$

Furthermore, since:
$$m_0 = 1, \quad m_1 = 0,$$

it follows that:
$$V = \varphi(\mu) + \frac{\varphi''(\mu)}{2} m_2 + \frac{\varphi^{(3)}(\mu)}{6} m_3 + \frac{\varphi^{(4)}(\mu)}{24} m_4 + \dots$$

That is:
$$V = \varphi(\mu) + \frac{\varphi''(\mu)}{2} \mathbb{E}[(X - \mu)^2] + \frac{\varphi^{(3)}(\mu)}{6} \mathbb{E}[(X - \mu)^3] + \frac{\varphi^{(4)}(\mu)}{24} \mathbb{E}[(X - \mu)^4] + \dots$$

This expresses $V$ as a **central moment expansion** of $X$.

## 2.3 Interpretation of Expansion Terms

From the expansion:
$$V \approx \varphi(\mu) + \frac{1}{2}\varphi''(\mu)m_2 + \frac{1}{6}\varphi^{(3)}(\mu)m_3 + \frac{1}{24}\varphi^{(4)}(\mu)m_4 + \dots$$

where:
$$m_2 = \mathbb{E}[(X - \mu)^2], \quad m_3 = \mathbb{E}[(X - \mu)^3], \quad m_4 = \mathbb{E}[(X - \mu)^4],$$

Each term can be interpreted as follows:

---

### 1. Second-Order Term
$$\frac{1}{2}\varphi''(\mu)m_2$$

This represents **"Local Curvature $\times$ Variance."** Since the curvature of a call payoff is primarily concentrated near the strike price, this term is most significant when $\mu \approx K$. That is, **ATM (At-The-Money)** option prices are mainly influenced by variance. When $\mu$ is far from $K$, the contribution of this term diminishes.

### 2. Third-Order Term
$$\frac{1}{6}\varphi^{(3)}(\mu)m_3$$

This represents **"Local Asymmetry $\times$ Skewness."** This term reflects the impact of the distribution's asymmetry on the price. Therefore, it primarily determines the left-right asymmetry of prices across different strikes—effectively the direction of the **volatility skew**.

### 3. Fourth-Order Term
$$\frac{1}{24}\varphi^{(4)}(\mu)m_4$$

This represents **"High-Order Local Bending $\times$ Kurtosis."** This term reflects the influence of heavy tails in the distribution. If a distribution has fatter tails, option values for strikes far from the mean will be higher. Consequently, this term mainly affects the elevation of the wings on both sides of the **volatility smile**.

---

### Conclusion
The **call option price** can be understood as being jointly determined by variance, skewness, and kurtosis:

* **Variance** primarily affects the price level near the ATM point.
* **Skewness** primarily determines the tilt direction of the volatility curve.
* **Kurtosis** primarily determines the curvature of the wings at both ends.

## Stochastic Volatility Model Analysis

**Note:** If we strictly follow the formula in the diagram:
$$dX_t = \alpha \sigma_t X_t dW_{1,t}, \quad X_0 = 0$$
Then $X_t \equiv 0$, and the subsequent moment formulas would not hold. This is likely a typo. Usually, the model is understood as follows:

$$dX_t = \alpha \sigma_t dW_{1,t}, \quad d\sigma_t = \nu \sigma_t dW_{2,t}, \quad dW_{1,t}dW_{2,t} = \rho dt$$
with $X_0 = 0$ and $\sigma_0 = 1$.

---

### 1. Calculating Variance $v = \mathbb{E}[X^2(T)]$

By **Itô Isometry**:
$$X_T = \alpha \int_{0}^{T} \sigma_t dW_{1,t}$$

Therefore:
$$v = \mathbb{E}[X_T^2] = \alpha^2 \mathbb{E} \left[ \int_{0}^{T} \sigma_t^2 dt \right] = \alpha^2 \int_{0}^{T} \mathbb{E}[\sigma_t^2] dt.$$

Since:
$$d\sigma_t = \nu \sigma_t dW_{2,t}, \quad \sigma_0 = 1,$$
we obtain:
$$\sigma_t = \exp \left( \nu W_{2,t} - \frac{1}{2}\nu^2 t \right).$$

Thus:
$$\mathbb{E}[\sigma_t^2] = e^{\nu^2 t}.$$

Consequently:
$$v = \alpha^2 \int_{0}^{T} e^{\nu^2 t} dt = \alpha^2 \frac{e^{\nu^2 T} - 1}{\nu^2}.$$

If $\nu^2 T$ is small, using the approximation $e^x \approx 1 + x$, we get:
$$v \approx \alpha^2 T.$$

---

### 2. Expressing $\nu^2 T, \rho, \alpha^2 T$ in terms of $s, \kappa, v$

Given the following approximations for skewness ($s$) and kurtosis ($\kappa$):
$$s \approx 3\rho \nu \sqrt{T}, \quad \kappa \approx 3 + 4\nu^2 T(1 + 3\rho^2), \quad v \approx \alpha^2 T.$$

Let $x = \nu^2 T$. Then:
$$s^2 = 9\rho^2 x.$$

Simultaneously:
$$\kappa - 3 = 4x(1 + 3\rho^2) = 4x + 12\rho^2 x.$$

Substituting $\rho^2 x = s^2 / 9$, we get:
$$\kappa - 3 = 4x + \frac{4}{3}s^2.$$

Therefore:
$$\nu^2 T = x = \frac{\kappa - 3}{4} - \frac{s^2}{3}.$$

Then, from $s = 3\rho \sqrt{x}$, we obtain:
$$\rho = \frac{s}{3\sqrt{\nu^2 T}} = \frac{s}{3\sqrt{\frac{\kappa - 3}{4} - \frac{s^2}{3}}}.$$

Finally:
$$\alpha^2 T \approx v.$$

---

### 3. Why this model fits the market better than BS / Bachelier

This model has **three degrees of freedom**:
* **$\alpha$** controls the variance level (volatility);
* **$\rho$** controls the skewness;
* **$\nu$** controls the kurtosis and the strength of the volatility smile.

As a result, it can simultaneously generate:
* Volatility levels;
* Skew;
* Fat tails / Kurtosis.
